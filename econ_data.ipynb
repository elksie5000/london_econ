## Working with London Datastore data in python

#### Just trying to work out how to quickly visualise a new dataset using a simple import of data from the web

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("https://data.london.gov.uk/download/london-economy-today/35e2fdda-2efa-447c-9e9a-b321205b14b2/LET%20Datastore%20-%20LM.xlsx", sheet_name="1")
df.head(30)
```

```python
column_names = df.loc[5, :].values.tolist()
```

```python
df = df.loc[6:373, :]
```

```python
df
```

```python
df.columns = column_names
df
```

```python
df2 =df.set_index("Quarter")
```

```python
 # Plot the data against the dataframe's index.
# For example, the London series is actually plotted as:
# (6, 0.12305), (7, 0.123895), (8, 0.124076), ...
ax = df[["London", "UK"]].plot()
# Tick the x-axis the same as the index, i.e: 6, 7, 8..., not 6, 6.5, 7, 7.5...
ax.xaxis.set_ticks(df.index)
# Label the ticks
#ax.xaxis.set_ticklabels(df["Year"])
plt.show()
```

```python
df2 = df2.dropna(axis=1)
```

```python
ax = df2.plot()
# rotate xlabels by 45 degrees
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```
